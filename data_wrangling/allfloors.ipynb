```python
import pandas as pd
import numpy as np
```

```python
# read all the datasets
```

```python
df_preconst = pd.read_csv('ParadiseView/clean/cost_to_preconstruction.csv')
```

```python
df_1stfloor = pd.read_csv('ParadiseView/clean/floorOne.csv')
```

```python
df_2ndfloor = pd.read_csv('ParadiseView/clean/floorTwo.csv')
```

```python
df_3rdfloor = pd.read_csv('ParadiseView/clean/FloorThree.csv')
```

```python
# combine all the datasets using concat method
df_paradiseView = pd.concat([df_preconst, df_1stfloor, df_2ndfloor, df_3rdfloor], ignore_index=True)
```

```python
df_paradiseView.info()
```

```python
# convert date column into datetime object
df_paradiseView['date'] = pd.to_datetime(df_paradiseView['date'])
```

```python
df_paradiseView.info()
```

```python
# print the entire dataset on screen
print(df_paradiseView.to_string())
```

```python
# save the full data as an excel file
df_paradiseView.to_excel('ParadiseView/clean/df_paradiseView.xlsx', index=False)
```

```python
# save the full dataset as a csv file
df_paradiseView.to_csv('ParadiseView/clean/df_paradiseView.csv', index=False)
```
