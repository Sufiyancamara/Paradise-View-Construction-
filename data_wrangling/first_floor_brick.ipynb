```python
# import pandas and numpy libraries
import pandas as pd
import numpy as np
```

```python
# declare two variables
filepath = "ParadiseView/Paradise View Expenditures 4.xlsx"
outFilePath = 'ParadiseView/out',
```

```python
# read an excel file into f_floorMolding
f_floorMolding = pd.read_excel(filepath, sheet_name='Brick Molding', skiprows=4)
```

```python
# display the dataset
f_floorMolding
```

```python
# drop four columns from the dataset
f_floorMolding.drop(columns=['Unnamed: 6', 'Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9'], inplace=True)
```

```python
f_floorMolding
```

```python
# keep rows with atleast six non-missing columns and drop the rest
f_floorMolding.dropna(thresh=len(f_floorMolding.columns) - 1, inplace=True)
```

```python
f_floorMolding
```

```python
# drop a row from the dataset where ITEM column is equal to ITEM
f_floorMolding.drop(f_floorMolding[f_floorMolding['ITEM'] == 'ITEM'].index, inplace =True)
```

```python
# make the dataset's columns lowercase strings
f_floorMolding.columns = f_floorMolding.columns.str.lower()
```

```python
# display first five rows of the dataset
f_floorMolding.head()
```

```python
# rename some of the columns
f_floorMolding.rename(columns={'unit price(d)': 'unit_cost', 'total cost(d)': 'total_cost'}, inplace=True)
```

```python
# drop cost($) column from the dataset
f_floorMolding.drop(columns='cost($)', inplace=True)
```

```python
f_floorMolding
```

```python
# aacess indeces and assign a value string to them
f_floorMolding.at[22, 'date'] = '2022-04-18 00:00:00'
```

```python
f_floorMolding.loc[32:, 'date'] = '2022-04-23 00:00:00'
```

```python
f_floorMolding.loc[f_floorMolding['item'] == 'Missing', 'date'] = 1
```

```python
f_floorMolding.at[32, 'unit_cost'] = 6.00
```

```python
f_floorMolding.at[22, 'unit_cost'] = 1500
```

```python
f_floorMolding.at[39, 'quantity '] = 1
f_floorMolding.at[39, 'unit_cost'] = 15
```

```python
# get the column names
f_floorMolding.columns
```

```python
# change the date column to datetime object
f_floorMolding['date'] = pd.to_datetime(f_floorMolding['date'])
```

```python
f_floorMolding['date'].info
```

```python
# change the format of the date column
f_floorMolding['date'] = f_floorMolding['date'].dt.strftime('%m/%d/%Y')
```

```python
f_floorMolding
```

```python
# create two new columns
f_floorMolding['sub_category'] = 'Brick Molding'
f_floorMolding['category'] = '1st Floor'
```

```python
f_floorMolding
```

```python
# save the dataset as a csv file
f_floorMolding.to_csv('ParadiseView/out/df_ffBrickMoldingClean.csv', index=False)
```
