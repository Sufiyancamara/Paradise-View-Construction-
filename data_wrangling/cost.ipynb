```python
import pandas as pd
import numpy as np
```

```python
# list of rows tobe dropped
skiprows = [0,1,2,3,6,7,8,9,10,22,23,24,25]
```

```python
# read an excel file into cost_date
cost_data = pd.read_excel("ParadiseView/Paradise View Expenditures 4.xlsx", skiprows=skiprows)
```

```python
cost_data
```

```python
# rename some of the columns 
cost_data.rename(columns={'COST(D)': 'cost_dalasi', 'COST($) ': 'cost_dollar'}, inplace=True)
```

```python
cost_data
```

```python
# change the columns names to lowercase string
cost_data.columns = cost_data.columns.str.lower()
```

```python
# change date column's datatype to datetime object
cost_data['date'] = pd.to_datetime(cost_data['date'])
```

```python
# change the format of the date column to month/day/year
cost_data['date'] = cost_data['date'].dt.strftime("%m/%d/%Y")
```

```python
cost_data
```

```python
# drop the cost_dollar column as it is not needed
df_cost = cost_data.drop(columns='cost_dollar')
```

```python
df_cost
```

```python
# rename cost_dalasi column to cost
df_cost.rename(columns={'cost_dalasi': 'cost'}, inplace=True)
```

```python
# save the dataset as aa csv file
df_cost.to_csv('ParadiseView/out/df_cost_clean.csv', index=False)
```
