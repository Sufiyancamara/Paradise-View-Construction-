```python
import pandas as pd
import numpy as np
```

```python
# read an excel file 
otherExp = pd.read_excel("ParadiseView/Paradise View Expenditures 4.xlsx", skiprows=55, sheet_name="Fencing")
```

```python
otherExp
```

```python
# drop unwanted columns from the dataset
otherExp.drop(['Unnamed: 4', 'Unnamed: 5', 'Unnamed: 6', 'Unnamed: 7', 'Unnamed: 8',], axis=1, inplace=True)
```

```python
# display the dataset
otherExp
```

```python
# drop all missing values from tghe dataset
otherExp.dropna(inplace=True)
```

```python
otherExp
```

```python
# rename columns
otherExp.rename(columns={'WHO': 'item', 'AMOUNT(D)': 'unit_cost'}, inplace=True)
```

```python
otherExp
```

```python
otherExp.columns
```

```python
# drop AMOUNT($) column from the dataset
otherExp.drop('AMOUNT($)', axis=1, inplace=True)
```

```python
otherExp
```

```python
# change column names to lowercase string
otherExp.columns = otherExp.columns.str.lower()
```

```python
# format the date column
otherExp['date'] = otherExp['date'].dt.strftime('%m/%d/%Y')
```

```python
otherExp
```

```python
# assign integer 1 to 'quantity' column
otherExp['quantity'] = 1
```

```python
otherExp
```

```python
# drive new column from results from multiplying two columns
otherExp['total_cost'] = otherExp['unit_cost'] * otherExp['quantity']
```

```python
# reindex the columns
otherExp = otherExp.reindex(columns=['item', 'quantity', 'unit_cost', 'total_cost', 'date'])
```

```python
otherExp
```

```python
# sdave the dataset as csv file without the indeces
otherExp.to_csv('ParadiseView/out/df_otherExp_clean.csv', index=False)
```
