```python
import pandas as pd
import numpy as np
```

```python
thirdfloor_bricks = pd.read_csv('ParadiseView/Paradise View/3rd_4th Floorbricks.csv')
```

```python
thirdfloor_bricks.drop(columns=(thirdfloor_bricks.loc[:, 'Unnamed: 6':'Unnamed: 11']), inplace=True)
```

```python
thirdfloor_bricks.info()
```

```python
thirdfloor_bricks = thirdfloor_bricks.dropna(thresh=3)
```

```python
thirdfloor_bricks.columns.values
```

```python
thirdfloor_bricks.columns = thirdfloor_bricks.columns.str.lower()
```

```python
thirdfloor_bricks.columns
```

```python
thirdfloor_bricks.rename(columns={'quality ': 'quantity', 'unit price(d)': 'unit_cost', 'cost(d)': 'total_cost', 'floor': 'category'}, inplace=True)
```

```python
thirdfloor_bricks.columns
```

```python
thirdfloor_bricks['date'] = pd.to_datetime(thirdfloor_bricks['date'])
```

```python
df_thirdfloor_bricks = thirdfloor_bricks[thirdfloor_bricks['date'] <= pd.to_datetime('2025-10-29')]
```

```python
df_fourthfloor_bricks = thirdfloor_bricks[thirdfloor_bricks['date'] > pd.to_datetime('2025-10-29')]
```

```python
df_fourthfloor_bricks
```

```python
df_thirdfloor_bricks
```

```python
df_thirdfloor_bricks['category'] = '3rd Floor'
```

```python
df_thirdfloor_bricks['total_cost'] = df_thirdfloor_bricks['total_cost'].str.replace(',', '').astype('float64')
```

```python
df_thirdfloor_bricks['unit_cost'] = df_thirdfloor_bricks['unit_cost'].str.replace(',', '').astype('float64')
```

```python
df_thirdfloor_bricks['quantity'] = df_thirdfloor_bricks['quantity'].str.replace(',', '').astype('float64')
```

```python
df_thirdfloor_bricks['sub_category'] = 'Brick Molding'
```

```python
df_thirdfloor_bricks = df_thirdfloor_bricks.reindex(columns=['item', 'quantity', 'unit_cost', 'total_cost', 'date', 'sub_category', 'category'])
```

```python
df_thirdfloor_bricks.info()
```

```python
print(df_thirdfloor_bricks.to_string())
```

```python
df_thirdfloor_bricks.to_csv('ParadiseView/out/df_thirdfloor_bricks.csv', index=False)
```
