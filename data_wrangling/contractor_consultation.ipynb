```python
import numpy as np
import pandas as pd
from datetime import datetime, date
```

```python
# Read the data from excel sheet
contractor_consul = pd.read_excel('ParadiseView/Paradise View Expenditures 4.xlsx', sheet_name='Consultation', skiprows= 4)
```

```python
contractor_consul
```

```python
# drop two unwanted columns
contractor_consul.drop(columns=['Unnamed: 6', 'Unnamed: 7'], inplace=True)
```

```python
# drop the cost($) column
contractor_consul.drop(columns='COST($)', inplace=True)
```

```python
contractor_consul
```

```python
# Permanentely drop missing data and keep rows with atmost 4 non-missing columns
contractor_consul.dropna(thresh = (len(contractor_consul.columns) - 2), axis=0, inplace=True)
```

```python
# pass a list of indeces to be dropped fron the dataset
contractor_consul = contractor_consul.drop(index=[2,9,12,21,24])
```

```python
contractor_consul
```

```python
# change the column names to lowecase and rename them to desire column names
contractor_consul.columns =contractor_consul.columns.str.lower()
contractor_consul.rename(columns={'quantity ': 'quantity', 'unit price(d)': 'unit_cost', 'total cost(d)': 'total_cost'}, inplace=True)
```

```python
contractor_consul.loc[22, 'date'] ='2023-12-20 00:00:00'
```

```python
# access and change values of columns to desire values
contractor_consul.loc[22, 'quantity'] = 1
contractor_consul.loc[22, 'unit_cost'] = 2000
contractor_consul.loc[22, 'total_cost'] = 2000
```

```python
contractor_consul
```

```python
contractor_consul['date'] = pd.to_datetime(contractor_consul['date'])
```

```python
contractor_consul['date'] = contractor_consul['date'].dt.strftime('%m/%d/%Y')
```

```python
# create three new collumns
contractor_consul['sub_category'] = 'Contractor Consultation'
contractor_consul['category'] = 'Preconstruction'
contractor_consul['exchange_rate'] = 61.50
```

```python
contractor_consul.loc[1:, 'exchange_rate'] = 64.00
```

```python
contractor_consul
```

```python
# change the date column to datatime object
contractor_consul['date'] = pd.to_datetime(contractor_consul['date'])
```

```python
# save the file as a csv file
contractor_consul.to_csv('ParadiseView/out/df_contractor_consul.csv', index=False)
```
