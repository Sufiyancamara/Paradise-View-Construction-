```python
# import pandas and numpy libraries
import pandas as pd 
import numpy as np
```

```python
# read an excel file into df_fencing
df_fencing = pd.read_excel("ParadiseView/Paradise View Expenditures 4.xlsx", skiprows=6, sheet_name="Fencing")
```

```python
# display the dataset
df_fencing
```

```python
# drop from the dataset from index number 40 to the end of the dataset
df_fencing = df_fencing.drop(df_fencing.iloc[40:].index)
```

```python
df_fencing
```

```python
# drop column 6 all the way to the last column
df_fencing = df_fencing.drop(df_fencing.iloc[:, 6:], axis=1)
```

```python
df_fencing
```

```python
# access row 25, column 1 and assign integer 4 to it
df_fencing.iloc[25:26, 1:2] = 4
```

```python
df_fencing.iloc[15:16, 1:2] = 1
```

```python
df_fencing
```

```python
# drop all rows with missing value
df_fencing.dropna(inplace=True)
```

```python
df_fencing
```

```python
# change the dataset column names into lowercase string
df_fencing.columns = df_fencing.columns.str.lower()
```

```python
# display the collumn names
df_fencing.columns
```

```python
# rename some column names
df_fencing.rename(columns={'unit price(d)': 'unit_cost', 'total cost(d)': 'total_cost', 'total($)': 'total_dollar'}, inplace=True)
```

```python
df_fencing
```

```python
# change date collumn format
df_fencing['date'] = df_fencing['date'].dt.strftime('%m/%d/%Y')
```

```python
df_fencing
```

```python
# drop the total_dollar column as it is not needed
df_fencing = df_fencing.drop('total_dollar', axis=1)
```

```python
df_fencing
```

```python
# save the datgaset as a csv file
df_fencing.to_csv('ParadiseView/out/df_fencing_clean.csv', index=False)
```
