```python
import numpy as np
import pandas as pd
```

```python
borehole = pd.read_excel('Paradiseview/Paradise View Expenditures 4.xlsx', sheet_name='Borehole', skiprows=3)
```

```python
borehole.head()
```

```python
borehole.dropna(thresh=3, axis=1, inplace=True)
```

```python
borehole
```

```python
borehole.dropna(thresh=(len(borehole.columns) - 1), axis=0, inplace=True)
```

```python
# Change the columns strings to lowercase unsing String Vectorization
borehole.columns = borehole.columns.str.lower()
```

```python
borehole.columns
```

```python
borehole = borehole.rename(columns={'unit price(d)': 'unti_cost', 'total cost(d)': 'total_cost'})
```

```python
# Perminantly drop the cost($) column fromn the dataset
borehole.drop(columns='cost($)', inplace=True)
```

```python
borehole
```

```python
# Chabge the misinput date to correct date
borehole.at[9, 'date'] = '2023-06-25'
```

```python
borehole
```

```python
borehole['date'].dtype
```

```python
# Check the datatype of date column
np.dtype(borehole['date'])
```

```python
borehole['date'] = borehole['date'].dt.strftime('%m/%d/%Y')
```

```python
# Assign 'Residual to index #9 from the item column
borehole.at[9, 'item'] = 'Residual'
```

```python
borehole.tail()
```

```python
# Create two new columns

borehole['sun_category'] = 'Borehole Drilling'
borehole['category'] = 'Preconstruction'
```

```python
borehole.rename(columns={'sun_category': 'sub_category'}, inplace=True)
```

```python
borehole.head()
```

```python
# Save the dataFrame to csv file
borehole.to_csv('ParadiseView/out/df_borehole_drilling.csv', index=False)
```
